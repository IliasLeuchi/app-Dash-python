```python
### Importation des données

## Arceaux
df_arceaux = gpd.read_file("arceaux.json")
## Agence
df_agence = gpd.read_file("agence_metro.json")
## Aire
df_aire = gpd.read_file("velo_service.json")
## Stationnement
df_stationnement =  gpd.read_file("stationnement_cyclable.json")
## Piste
warnings.filterwarnings('ignore')
df_piste_gre  = gpd.read_file("piste_metro.json")
df_piste_voiron =  gpd.read_file("piste_voiron.json")
piste_voiron  = df_piste_voiron.explode()
df_piste_gresi = gpd.read_file("piste_gresivaudan.json")

## Accident

# Caractéristique
df_carac_2015 =pd.read_csv("data_caracteristiques/caracteristiques-2015.csv")
df_carac_2016 = pd.read_csv("data_caracteristiques/caracteristiques-2016.csv")
df_carac_2017 = pd.read_csv("data_caracteristiques/caracteristiques-2017.csv")
df_carac_2018 = pd.read_csv("data_caracteristiques/caracteristiques-2018.csv")
df_carac_2019 =pd.read_excel("data_caracteristiques/caracteristiques-2019.xlsx")
df_carac_2020 =  pd.read_excel("data_caracteristiques/caracteristiques-2020.xlsx")
# Véhicules
df_vehicule_2015 = pd.read_csv("data_vehicules/vehicules-2015.csv")
df_vehicule_2016 =  pd.read_csv("data_vehicules/vehicules-2016.csv")
df_vehicule_2017 = pd.read_csv("data_vehicules/vehicules-2017.csv")
df_vehicule_2018 = pd.read_csv("data_vehicules/vehicules-2018.csv")
df_vehicule_2019 =  pd.read_csv("data_vehicules/vehicules-2019.csv", sep= ";")
df_vehicule_2020 = pd.read_csv("data_vehicules/vehicules-2020.csv", sep= ";")
```

```python
df_usager2015 = pd.read_csv("data_usagers/usagers-2015.csv")
df_usager2015["an"] = 2015
df_usager2016 = pd.read_csv("data_usagers/usagers-2016.csv")
df_usager2016["an"] = 2016
df_usager2017 = pd.read_csv("data_usagers/usagers-2017.csv")
df_usager2017["an"] = 2017
df_usager2018 = pd.read_csv("data_usagers/usagers-2018.csv")
df_usager2018["an"] = 2018
df_usager2019 = pd.read_csv("data_usagers/usagers-2019.csv",sep= ";")
df_usager2019["an"] = 2019
df_usager2020 = pd.read_csv("data_usagers/usagers-2020.csv",sep= ";")
df_usager2020["an"] = 2020

df_usager2019_2020 = pd.concat([df_usager2019, df_usager2020])
df_usager2019_2020 = df_usager2019_2020[["Num_Acc","grav","an"]]


df_usager = pd.concat([df_usager2015, df_usager2016,df_usager2017,df_usager2018])
df_usager = df_usager[["Num_Acc","grav","an"]]
df_usager = pd.concat([df_usager,df_usager2019_2020])
df_usager['grav'] = df_usager['grav'].replace(1, 'Indemne')
df_usager['grav'] = df_usager['grav'].replace(2, 'Tué')
df_usager['grav'] = df_usager['grav'].replace(3, 'Blessé hospitalisé')
df_usager['grav'] = df_usager['grav'].replace(4, 'Blessé léger')
```

```python
# Concatenate
df_carac = pd.concat([df_carac_2015,df_carac_2016,df_carac_2017,df_carac_2018])
df_carac["an"] = df_carac["an"]+2000
df_carac["lat"] = df_carac["lat"]*0.00001
df_carac["long"] = df_carac["long"]*0.00001

# Filtrer les données en fonction de la zone géographique de l'agglomération de Grenoble
df_carac = df_carac[(df_carac['lat'] >= lat_min) & (df_carac['lat'] <= lat_max) & 
                 (df_carac['long'] >= lon_min) & (df_carac['long'] <= lon_max)]

df_carac_19_20 = pd.concat([df_carac_2019, df_carac_2020])
df_carac_19_20 = df_carac_19_20[df_carac_19_20['com'].isin(com_gre)]

df_carac = pd.concat([df_carac,df_carac_19_20])
df_carac.drop(["gps", "jour","hrmn","int","col"],axis=1, inplace= True)

df_carac['agg'] = df_carac['agg'].replace(1, 'Hors agglomération')
df_carac['agg'] = df_carac['agg'].replace(2, 'En agglomération')

df_carac['lum'] = df_carac['lum'].replace(1, 'Plein jour')
df_carac['lum'] = df_carac['lum'].replace(2, 'Crépuscule ou aube')
df_carac['lum'] = df_carac['lum'].replace(3, 'Nuit sans éclairage public')
df_carac['lum'] = df_carac['lum'].replace(4, 'Nuit avec éclairage public non allumé ')
df_carac['lum'] = df_carac['lum'].replace(5, 'Nuit avec éclairage public allumé')

df_carac['atm'] = df_carac['atm'].replace(-1, 'Non renseigné')
df_carac['atm'] = df_carac['atm'].replace(1, 'Normale')
df_carac['atm'] = df_carac['atm'].replace(2, 'Pluie légère')
df_carac['atm'] = df_carac['atm'].replace(3, 'Pluie forte')
df_carac['atm'] = df_carac['atm'].replace(4, 'Neige-grêle')
df_carac['atm'] = df_carac['atm'].replace(5, 'Brouillard-fumée')
df_carac['atm'] = df_carac['atm'].replace(6, 'Vent fort-tempête')
df_carac['atm'] = df_carac['atm'].replace(7, 'Temps éblouissant')
df_carac['atm'] = df_carac['atm'].replace(8, 'Temps couvert')
df_carac['atm'] = df_carac['atm'].replace(9, 'Autre')

df_carac = df_carac[df_carac['dep'].isin(dep_gre)]
df_carac['nom_ville'] = df_carac['com'].map(communes)

df_vehicule_19_20 = pd.concat([df_vehicule_2019,df_vehicule_2020])
df_vehicule = pd.concat([df_vehicule_2015,df_vehicule_2016,df_vehicule_2017,df_vehicule_2018])
df_vehicule_19_20 = df_vehicule_19_20[["Num_Acc","catv"]]
df_vehicule = df_vehicule[["Num_Acc","catv"]]
df_vehicule  = pd.concat([df_vehicule,df_vehicule_19_20])
AccVelo = df_vehicule[df_vehicule["catv"] == 1]["Num_Acc"] 
Num_Acc_Velo = AccVelo.tolist()

df_AccVeloGre = df_carac[df_carac["Num_Acc"].isin(Num_Acc_Velo)]
Num_acc_vel_gre = df_AccVeloGre["Num_Acc"].tolist()
df_usager_gre = df_usager[df_usager["Num_Acc"].isin(Num_acc_vel_gre)]
Nb_grav = df_usager_gre.groupby(["grav","an"]).agg({"Num_Acc" : "count"})
Nb_grav.reset_index(inplace=True)
Nb_grav.rename(columns = {"Num_Acc": "nb_grav"}, inplace = True)

```

```python
app = JupyterDash(__name__)

 

#app.css.append_css({'external_url': external_stylesheets})


### Définition de variables
colors = {
    'yellow': '#FFEC44',
    'black': '#000000',
    'white': '#FFFFFF',
    'blue': '#353967',
    'red': '#e84626',
    'text': '#7FDBFF' 
}

#--------------------------------------------------------------------------------------------------------------------------

### App
app.layout = html.Div(style={'backgroundColor': colors['yellow'], 'border-left': 'solid 80px #353967', 
                             'border-right': 'solid 80px #353967', 'border-top': 'solid 30px #353967', 
                             'border-bottom': 'solid 30px #353967', 'border-radius': '30px'}, children = [
    html.H1(
        children = "METRO CYCLO SECU",
        style = { 'textAlign': 'center',
                    'color': '#353967',
                    "border" : '#353967',
                    "font-family": "arial",
                }
    ),
    
#--------------------------------------------------------------------------------------------------------------------------

    html.H2(
        children = "Infrastructure",
        style = { 'textAlign': 'center',
                    'color': colors['yellow'],
                    "font-family": "arial",
                     'border': 'solid 5px #353967',
                     "backgroundColor" : "#353967"
                }
    ),

  html.Div(
    children = [
        # Radio Button
        html.Div([
            
            html.Label("Afficher les aires de services ?"),
            dcc.RadioItems(
                id='aires', 
                options=["Oui", "Non"],
                value="Non",
                inline=True
            ),
        ],style={'order': '1', 'flex-basis': '16%', 'padding': '10px', "font-family": "arial"}),
        html.Div([
            html.Label("Afficher les arceaux ?"),
            dcc.RadioItems(
                id='arceaux', 
                options=["Oui", "Non"],
                value="Non",
                inline=True
            ),
        ],style={'order': '2', 'flex-basis': '15%', 'padding': '10px', 
                 "font-family": "arial"}),
        html.Div([
            html.Label("Afficher les stationnements ?"),
            dcc.RadioItems(
                id='stationnements', 
                options=["Oui", "Non"],
                value="Non",
                inline=True
            ),
        ], style={'order': '3', 'flex-basis': '18%', 'padding': '10px',
                  "font-family": "arial"}),
        
        # Dropdown
        
        html.Div([
             dcc.Dropdown(
                id='communes-dropdown',
                options=[{'label': c, 'value': c} for c in communes],
                value=[communes[0], communes[11], communes[6]],
                multi=True)
            ], style={'order': '4', 'flex-basis': '50%', 'padding': '10px',
                      "font-family": "arial"}
        ),
    ], style={'display': 'flex', 'padding': '10px'}
  ),
    
#--------------------------------------------------------------------------------------------------------------------------
       
   html.Div(
    children = [
        
        #Map infrastructure 
        html.Div(
            children = [
                dcc.Graph(id="map",
                          style= {"border" : "10px solid #353967"}),
                
            ], style={'order': '1', 'flex-basis': '50%', 'padding': '10px',
                      "font-family": "arial"}
        ),
        
        # Graphique 
         html.Div(
            children = [
                dcc.Graph(id='stationnement-graph', 
                        style= {"border" : "10px solid #353967"})
                
            ], style={'order': '2', 'flex-basis': '50%', 'padding': '10px', 
                      "font-family": "arial"}
        ),  
    ], style={'display': 'flex', 'padding': '10px' ,"font-family": "arial"}
),
    
#--------------------------------------------------------------------------------------------------------------------------

    html.Div(
        children = [

            #Barplot Arceaux 
            html.Div(
                children = [
                    dcc.Graph(id="figArceaux", figure = figArceaux , 
                              style= {"border" : "10px solid #353967"}),
                    html.Label("Une année non présente signifie qu'aucun arceau à été installé.",
                              style={"font-style": "italic"}),
                ], style={'order': '1', 'flex-basis': '50%', 'padding': '10px',
                          "font-family": "arial"}
            ),

            # Boxplot 
             html.Div(
                children = [
                    dcc.Graph(id="Boxplot", figure = boxplot, 
                              style= {"border" : "10px solid #353967"}),
                    html.Label("Ce graphique aide à choisir le type de stationnement à installé en fonction du nombre de places qu'on désir mettre en place."
                              ,style={"font-style": "italic"}),
                ], style={'order': '2', 'flex-basis': '50%', 'padding': '10px',
                          "font-family": "arial"}
            ),
        ], style={'display': 'flex', 'padding': '10px',"font-family": "arial"}
    ),
#--------------------------------------------------------------------------------------------------------------------------
   html.H2(
            children = "Accidents à vélo",
       style = { 'textAlign': 'center',
                    'color': colors['yellow'],
                    "font-family": "arial",
                     'border': 'solid 5px #353967',
                     "backgroundColor" : "#353967"
                }
        ),
    
    ## RangeSlider
    html.Div(
        children = [
            dcc.RangeSlider(
                min = 2015,
                max = 2020,
                step = 1,
                marks = {
                    2015: '2015',
                    2016: '2016',
                    2017: '2017',
                    2018: '2018',
                    2019: '2019',
                    2020: '2020',
                },
                value = [2015,2020],
                id = "slider_year",
                
            ),
        ], 
    ),
    
    
    html.Div(
        children = [
            ## Map Accident
            html.Div(
                
                children = [
                    dcc.Graph(id="map_acci", 
                              style= {"border" : "10px solid #353967"}),
                   
                ], style={'order': '1', 'flex-basis': '50%', 'padding': '10px',
                          "font-family": "arial", 'border-radius': 'solid 30px'}
            ),



            ## piechart
            html.Div(
                children = [
                    dcc.Graph(id = "piechart", 
                              style= {"border" : "10px solid #353967"}),
                    html.Label("Vous pouvez cliquer sur un type d'accident pour filtrer la carte à gauche."
                              ,style={"font-style": "italic"})
                    
                ], style={'order': '2', 'flex-basis': '50%', 'padding': '10px', 
                          "font-family": "arial", 'border-radius': 'solid 30px'}
            ),


        ], style={'display': 'flex', 'padding': '10px', "font-family": "arial"}
    ),
    
    
    html.Div(
            ## Lineplot
                children = [
                    dcc.Graph(id="lineplot", figure = line, 
                              style= {"border" : "10px solid #353967"}),
                   
                ],style={'padding': '10px', "font-family": "arial"}
    ),


#--------------------------------------------------------------------------------------------------------------------------
])

###################################################################################################################################
###################################################################################################################################

### Callbacks
@app.callback(
    Output("map", "figure"),
    Input("aires", "value"),
    Input("arceaux", "value"),
    Input("stationnements", "value"))

def piste_map(aires, arceaux, stationnements):
    ## Piste
    # Piste grenoble
    lats = []
    lons = []   
    for linestring in df_piste_gre.geometry:
        x, y = linestring.xy
        lats = np.append(lats, y)
        lons = np.append(lons, x)
        lats = np.append(lats, None)
        lons = np.append(lons, None)    
    fig1 = px.line_mapbox(lat=lats, lon=lons,
        mapbox_style="carto-positron", 
        zoom=11,
        title = "Réseaux cyclable")

    # Piste grésivaudan
    lats = []
    lons = []
    for linestring in df_piste_gresi.geometry:
        x, y = linestring.xy
        lats = np.append(lats, y)
        lons = np.append(lons, x)
        lats = np.append(lats, None)
        lons = np.append(lons, None)
    fig1_2 = px.line_mapbox(lat=lats, lon=lons,
        mapbox_style="carto-positron", 
        zoom=11)
    # Piste Voiron
    lats = []
    lons = []
    for linestring in piste_voiron.geometry:
        x, y = linestring.xy
        lats = np.append(lats, y)
        lons = np.append(lons, x)
        lats = np.append(lats, None)
        lons = np.append(lons, None)
    fig1_3 = px.line_mapbox(lat=lats, lon=lons,
        mapbox_style="carto-positron", 
        zoom=11,
        #title = ""
             )
    
    
    # Piste
    fig1.add_trace(fig1_2.data[0])
    fig1.add_trace(fig1_3.data[0])
    fig1.update_layout(margin={"r":10,"t":27,"l":10,"b":10})

    # Arceaux
    fig2 = px.scatter_mapbox(df_arceaux, lat= df_arceaux.geometry.y , lon= df_arceaux.geometry.x ,
        mapbox_style = "carto-positron",
        custom_data = ["annee"],
        #title = "",
        zoom = 11,
        color_discrete_sequence = [colors["red"]]
        )
    fig2.update_traces(hovertemplate="<br>".join(["Année de création : %{customdata[0]}"])),

    # Aires de service
    fig3 = px.scatter_mapbox(df_aire ,lat= df_aire.geometry.y , lon= df_aire.geometry.x,
                       mapbox_style = "carto-positron",
                       custom_data = ["pose"],
                       title = "Carte des services de vélo de la métro Grenoble",
                       zoom = 11,
                       color_discrete_sequence=["yellow"]
                       )

    fig3.update_traces(hovertemplate="<br>".join(["Années de pose : %{customdata[0]}"])),
    
    # Stationnement
    fig4 = px.scatter_mapbox(df_stationnement, lat= "ylat" , lon="xlong",
                       mapbox_style = "carto-positron",
                       custom_data = ["commentaires"],
                       zoom = 11,
                       color_discrete_sequence=["orange"]
                       )

    fig4.update_traces(hovertemplate="<br>".join(["type de stationnement : %{customdata[0]}"]))

    # Map finale
    if arceaux == "Oui":
        if aires == 'Oui':
            if stationnements == 'Oui':
                fig = fig1.add_trace(fig2.data[0])
                fig.add_traces(fig3.data[0])
                fig.add_traces(fig4.data[0])
            else:
                fig = fig1.add_trace(fig2.data[0])
                fig.add_traces(fig3.data[0])
        else: 
            if stationnements == 'Oui':
                fig = fig1.add_trace(fig2.data[0])
                fig.add_traces(fig4.data[0])
            else:
                fig = fig1.add_trace(fig2.data[0])

    else: 
        if aires == 'Oui':
            if stationnements == 'Oui':
                fig = fig1.add_trace(fig3.data[0])
                fig.add_traces(fig4.data[0])
            else:
                fig = fig1.add_trace(fig3.data[0])
        else: 
            if stationnements == 'Oui':
                fig = fig1
                fig.add_trace(fig4.data[0])
            else:
                fig = fig1

# add_trace
    return fig

#--------------------------------------------------------------------------------------------------------------------------

@app.callback(
    Output("map_acci", "figure"),
    Input("slider_year", "value"),
    Input("piechart", "clickData"))


def map_acci(slider_year, grav_click):
    if grav_click is None:
        fig_acci = px.scatter_mapbox(df_AccVeloGre[df_AccVeloGre["an"].between(min(slider_year),max(slider_year))],
            lat= "lat", 
            lon= 'long',
            mapbox_style = "carto-positron",
            custom_data = ["an"],
            title = "Accidents de vélo",
            zoom = 11,
            color = 'grav',
            labels = {"grav": "Gravité de l'accident"})
    
    else:
        selected = str(grav_click['points'][0]['label'])
        if selected == 'Tué':
            col = px.colors.qualitative.Plotly[0]
        elif selected == 'Indemne':
            col = px.colors.qualitative.Plotly[2]
        elif selected == 'Blessé léger':
            col = px.colors.qualitative.Plotly[1]
        elif selected == 'Blessé hospitalisé':
            col = px.colors.qualitative.Plotly[3]
        df_map = df_AccVeloGre[df_AccVeloGre['grav'] == selected]
        fig_acci = px.scatter_mapbox(df_map[df_map["an"].between(min(slider_year),max(slider_year))],
            lat= "lat", 
            lon= 'long',
            mapbox_style = "carto-positron",
            custom_data = ["an"],
            title = "Accidents de vélo",
            zoom = 11,
            color = 'grav',
            color_discrete_sequence=[col],
            labels = {"grav": "Gravité de l'accident"})
    
    fig_acci.update_traces(hovertemplate="<br>".join(["Année : %{customdata[0]}"]))
    fig_acci.update_layout(margin={"r":10,"t":27,"l":10,"b":10})
    
    return fig_acci

#--------------------------------------------------------------------------------------------------------------------------

# Fonction de rappel pour mettre à jour le graphique en fonction des communes sélectionnées
@app.callback(
    Output('stationnement-graph', 'figure'),
    Input('communes-dropdown', 'value')
)
def update_graph(communes_selectionnees):
    # Filtrer les données pour n'inclure que les communes sélectionnées
    data = Capacite_ville[Capacite_ville['commune_local'].isin(communes_selectionnees)]
    # Créer le graphique avec les données filtrées
    fig1 = px.bar(data.sort_values('capacite', ascending=True),
                 x='capacite', y='commune_local',  text= "capacite",
                 title="Capacité de stationnement par communes", 
                    labels = {'commune_local': 'Communes', 'capacite': "Capacité"},
                  color_discrete_sequence=["orange"])
    
    return fig1

#--------------------------------------------------------------------------------------------------------------------------

@app.callback(
    Output('piechart', 'figure'),
    Input('slider_year', 'value'))

def update_piechart(selected_years):
    if selected_years[0] == selected_years[1] :
        t="Proportion des gravités des accidents : "+str(selected_years[0])
    else :
        t="Proportion des gravités des accidents : "+str(selected_years[0])+" - "+str(selected_years[1])
                 
    filtered_df = Nb_grav[(Nb_grav['an'] >= selected_years[0]) & (Nb_grav['an'] <= selected_years[1])]
    pie = px.pie(filtered_df, values='nb_grav', names='grav',
                 color = 'grav',
                 #color_discrete_map = {'Tué':Plotly[0],'Blessé léger':4,'Blessé hospitalisé':4,'Indemme':1},
                 color_discrete_sequence=[px.colors.qualitative.Plotly[3],px.colors.qualitative.Plotly[1],px.colors.qualitative.Plotly[2],px.colors.qualitative.Plotly[0]],
                 labels = {'grav': 'Gravité', 'nb_grav': "Nombre"},
                 title = t,
                 hole=.3)
    pie.update_layout(clickmode='event+select')
                 
    return pie





```
